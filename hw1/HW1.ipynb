#### Q2

The more complex a model is, the more data points it will capture, and the lower the bias will be. However, if a model is too complex, when adding new points, the model can change a lot, and the predicted value will also change greatly, therefore causing a large variance. But when a model is simple, it will not catch all points and have a large bias, but the variance can be small.

Regularization contols overfitting and underfitting by minimizing the residual sum of squares after penalization. α controls the size of coefficient and the amount of regularization. If α is too large, the model will tend to be underfitting and have high MSE, if α is too small, the model will tend to be overfitting and have high MSE as well. So a suitable α will lead to a small MSE. Take Q1 for example, the coefficients of some variables are shrunk to a value close to 0. In this way, unimportant variables' affects will be reduced, and the model will be less likely to overfit.
